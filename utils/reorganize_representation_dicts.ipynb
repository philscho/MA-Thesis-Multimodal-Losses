```python
import os
import pickle
import numpy as np
from datetime import datetime

FOLDER = "/home/phisch/multimodal/data/representations/CIFAR10/full_dataset_aug_mlm"

def parse_filename(filename):
    # Example: CIFAR10-ckpt123-mean_class_reps-text_proj.pkl
    parts = filename.split("-")
    dataset = parts[0]
    model = parts[1]
    return dataset, model

for fname in os.listdir(FOLDER):
    if "mean_class_reps-vis_enc" in fname and fname.endswith(".pkl"):
        file_path = os.path.join(FOLDER, fname)
        with open(file_path, "rb") as f:
            arr = pickle.load(f)
        dataset, model = parse_filename(fname)
        arr = np.array(arr)
        # Build the new structure
        entry = {
            "dataset": dataset,
            "model": model,
            "class_names": ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"],
            "representations": [
                {
                    "modality": "vision",
                    "layer": "last_encoder_layer",
                    "data": arr,
                    "embedding_dim": arr.shape[1] if arr.ndim > 1 else None,
                    "dtype": str(arr.dtype),
                    "created": datetime.now().strftime("%Y-%m-%d"),
                    "notes": "",
                }
            ]
        }
        # Save the new dictionary to a file with a similar name, overwriting if it exists
        out_fname = fname.replace("mean_class_reps-vis_enc", "visual_representations")
        out_path = os.path.join(FOLDER, out_fname)
        if os.path.exists(out_path):
            with open(out_path, "rb") as f:
                data = pickle.load(f)
                data["representations"].append(entry["representations"][0])
                entry = data
        with open(out_path, "wb") as f:
            pickle.dump(entry, f)
        print(f"Overwritten: {out_path}")
```
