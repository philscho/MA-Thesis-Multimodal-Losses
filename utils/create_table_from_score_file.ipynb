```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/home/phisch/multimodal/test_results/model_scores_linear_probe.csv'
df = pd.read_csv(file_path)
```

```python
# Filter the DataFrame for rows where dataset_fraction is '1-aug'
filtered_df = df[(df['dataset_fraction'] == '1-aug') &
                 (df['metric'] == 'Top1Accuracy')]
```

```python
filtered_df
```

```python
# Pivot the table to create rows for model names and columns for datasets
pivot_table = filtered_df.pivot_table(
    index='model_name',  # Rows: model names
    columns='dataset',   # Columns: datasets
    values='score',  # Values: Top1Accuracy
    aggfunc='first'  # Use the first value if duplicates exist
)
```

```python
# Print the table
print(pivot_table)
```

```python
# Save the table to a CSV file (optional)
output_path = '/home/phisch/multimodal/test_results/linear_probe-top1acc-higher_aug_models.csv'
pivot_table.to_csv(output_path)
```
