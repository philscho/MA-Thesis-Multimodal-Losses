```python
import os
import pickle
from pathlib import Path
```

```python
templates = ["0_templates", "1_template", "5_templates", "18_templates"]
checkpoints = ["zathvtrx", "9v1wy0vb", "oc0g8fql", "9nvg456i", "93t3xgrr", "yh1adr3g"]

template_dict = {k: dict() for k in templates}
new_dicts = {k: {"zeroshot": template_dict} for k in checkpoints}
print(template_dict)
print(new_dicts)
```

```python
p = Path("test_results/zero_shot/full_dataset/")

for template in templates:
    tdir = os.path.join(p, template)
    flist = os.listdir(tdir)
    for fname in flist:
        model_id = fname.split("_")[0]
        fpath = os.path.join(tdir, fname)
        print(fpath)
        with open(fpath, "rb") as f:
            data = pickle.load(f)
            print(model_id, ":", template)
            new_dicts[model_id]["zeroshot"][template] = data
```

```python
for model_id in new_dicts:
    with open(f"test_results/zero_shot/full_dataset/{model_id}-zeroshot.pkl", "wb") as f:
        pickle.dump(new_dicts[model_id], f)
```
