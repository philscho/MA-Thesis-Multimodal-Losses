```python
import pickle
import os
from src.analysis.torch_cka.utils import plot_results

for fname in os.listdir("/home/phisch/multimodal/ddd"):
    if not fname.endswith(".pkl"):
        continue
    with open(f"/home/phisch/multimodal/ddd/{fname}", 'rb') as f:
        data = pickle.load(f)

    model2_name = data["model2_name"]
    model1_name = data["model1_name"]

    plot_results(
        hsic_matrix=data["CKA"],
        xlabel=model2_name,
        ylabel=model1_name,
        save_path=f"/home/phisch/multimodal/ddd/{fname}".replace(".pkl", ".png"),
        title=f"{model1_name} vs {model2_name}",
        )
```
